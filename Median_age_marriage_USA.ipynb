```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("Median_age_marriage_USA.csv")

# Rename columns to ensure they match expected names
df.columns = ["Year", "Median_Age_Men", "Median_Age_Women"]

# Check the first few rows of the dataset
print(df.head())
print(df.info())
```

```python
# Convert 'Year' to numeric, forcing errors to NaN (to handle cases like "2021r")
df["Year"] = pd.to_numeric(df["Year"])
```

```python
# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(df["Year"], df["Median_Age_Men"], marker="o", linestyle="-", label="Men")
plt.plot(df["Year"], df["Median_Age_Women"], marker="s", linestyle="-", label="Women")

# Customize the chart
plt.xlabel("Year")
plt.ylabel("Median Age at First Marriage")
plt.title("Median Age of Marriage in the U.S. (1947-2024)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```
