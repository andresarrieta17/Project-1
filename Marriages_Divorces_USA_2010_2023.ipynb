```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
df_2023 = pd.read_csv("Marital_status_USA_2023.csv")
df_2022 = pd.read_csv("Marital_status_USA_2022.csv")
df_2021 = pd.read_csv("Marital_status_USA_2021.csv")
df_2019 = pd.read_csv("Marital_status_USA_2019.csv")
df_2018 = pd.read_csv("Marital_status_USA_2018.csv")
df_2017 = pd.read_csv("Marital_status_USA_2017.csv")
df_2016 = pd.read_csv("Marital_status_USA_2016.csv")
df_2015 = pd.read_csv("Marital_status_USA_2015.csv")
df_2014 = pd.read_csv("Marital_status_USA_2014.csv")
df_2013 = pd.read_csv("Marital_status_USA_2013.csv")
df_2012 = pd.read_csv("Marital_status_USA_2012.csv")
df_2011 = pd.read_csv("Marital_status_USA_2011.csv")
df_2010 = pd.read_csv("Marital_status_USA_2010.csv")
```

```python
# Create a list of all dataframes
dataframes = [
    df_2023, df_2022, df_2021, df_2019, df_2018, df_2017, df_2016,
    df_2015, df_2014, df_2013, df_2012, df_2011, df_2010
]

# Assign the corresponding years to each dataframe
years = [2023, 2022, 2021, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010]

# Add the year column to each dataframe
for df, year in zip(dataframes, years):
    df["Year"] = year

# Concatenate all dataframes into one
merged_df = pd.concat(dataframes, ignore_index=True)

# Save the merged dataset
merged_filename = "Marital_status_USA_2010_2023.csv"
merged_df.to_csv(merged_filename, index=False)

# Display first few rows to confirm the merge
print(merged_df.head())

print(f"All files have been successfully merged and saved as {merged_filename}.")
```

```python
display(merged_df.info())
display(merged_df.head())
```

```python
# Define new column names
new_column_names = {
    "Label (Grouping)": "Demographic Group",
    "United States!!Total!!Estimate": "Total Population",
    "United States!!Total!!Margin of Error": "Total Population Margin of Error",
    "United States!!Now married (except separated)!!Estimate": "Married(%)",
    "United States!!Now married (except separated)!!Margin of Error": "Married Margin of Error",
    "United States!!Widowed!!Estimate": "Widowed(%)",
    "United States!!Widowed!!Margin of Error": "Widowed Margin of Error",
    "United States!!Divorced!!Estimate": "Divorced(%)",
    "United States!!Divorced!!Margin of Error": "Divorced Margin of Error",
    "United States!!Separated!!Estimate": "Separated(%)",
    "United States!!Separated!!Margin of Error": "Separated Margin of Error",
    "United States!!Never married!!Estimate": "Never_Married(%)",
    "United States!!Never married!!Margin of Error": "Never Married Margin of Error",
    "Year": "Year"
}

# Rename the columns
merged_df.rename(columns=new_column_names, inplace=True)

# Display the first few rows to confirm the changes
display(merged_df.head())
```

```python
display(merged_df.info())
```

```python
# Define percentage columns
percentage_columns = ["Married(%)", "Widowed(%)", "Divorced(%)", "Separated(%)", "Never_Married(%)"]

# Convert percentage columns to numeric (handling errors)
for col in percentage_columns:
    merged_df[col] = (
        merged_df[col]
        .astype(str)  # Ensure it's a string
        .str.replace("%", "", regex=False)  # Remove percentage signs
        .replace("(X)", np.nan)  # Replace "(X)" with NaN
        .replace("-", np.nan)  # Replace missing values with NaN
        .astype(float)  # Convert to float
    )

# Aggregate data by year (taking the mean across groups, ignoring NaNs)
yearly_data = merged_df.groupby("Year")[percentage_columns].mean()
```

```python
# Plot each marital status trend over time
plt.figure(figsize=(12, 6))
for col in percentage_columns:
    plt.plot(yearly_data.index, yearly_data[col], marker="o", linestyle="-", label=col)

# Customize the chart
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.title("Trends in Marital Status in the U.S. (2010-2023)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
```

```python
# Define percentage columns to use for the stacked area chart
percentage_columns = ["Married(%)", "Widowed(%)", "Divorced(%)", "Separated(%)", "Never_Married(%)"]

# Convert percentage columns to numeric
for col in percentage_columns:
    merged_df[col] = pd.to_numeric(
        merged_df[col].astype(str).str.replace("%", "", regex=False), errors="coerce"
    )

# Filter the dataset to only include females
df_female = merged_df[merged_df["Gender"] == "Female"]

# Get unique age groups
age_groups = df_female["Demographic Group"].unique()

# Create stacked area plots for each age group
for age_group in age_groups:
    # Filter the dataset for the specific age group
    df_age_group = df_female[df_female["Demographic Group"] == age_group]

    # Ensure the data is sorted by year
    df_age_group = df_age_group.sort_values(by="Year")

    # Prepare data for the stacked area chart
    years = df_age_group["Year"]
    data = df_age_group[percentage_columns]

    # Plot the stacked area chart
    plt.figure(figsize=(10, 6))
    plt.stackplot(
        years,
        data.T,  # Transpose data to stackplot format
        labels=["Married", "Widowed", "Divorced", "Separated", "Never Married"],
        alpha=0.8
    )

    # Add labels, title, and legend
    plt.title(f"Marital Status Distribution (Females, {age_group})")
    plt.xlabel("Year")
    plt.ylabel("Percentage")
    plt.legend(loc="upper left", title="Marital Status")
    plt.grid(True)

    # Show plot
    plt.tight_layout()
    plt.show()

```
